```python
import pandas as pd
```

```python
df = pd.read_csv('./gamedata_good.csv')
df.sample(15)
```

```python
df['Genre'] = df['Genre'].str.split(',').str[0]
df.head()
```

```python
df['Genre'].value_counts()
```

```python
other = ('Pinball', 'Educational', 'Compilation', 'Productivity', 'None',
        'Casino', 'Battle', 'Adult')
#Baseball           

df['Genre'] = df['Genre'].replace(['Pinball', 'Educational', 'Compilation', 'Productivity', 'None', 
                                   'Casino', 'Battle', 'Adult', 'Virtual Pet', 'Baseball'],
                                  ['Other', 'Other', 'Other', 'Other', 'Other', 'Other', 'Other',
                                   'Other', 'Other', 'Sports'])
                
df['Genre'].value_counts()
```

```python
def recommend(console, genre):
    system = df[df['Platform'] == console]
    games = system[system['Genre'] == genre]
    return games.sample(1)
```

```python
recommend('Xbox One', 'Shooter')
```
